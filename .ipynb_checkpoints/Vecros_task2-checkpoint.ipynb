```python
n = 10
grid_data = np.zeros((n*n*n,3))
l = 0
for i in range(n):
    for j in range(n):
        for k in range(n):
            grid_data[l] = np.array([i,j,k])
            l += 1

Grid = np.zeros((n,n,n))

np.random.seed(0)

def create_obstacle():
    obstacle_data = np.zeros((n*n,3))
    
    for i in range(10*10):
        x = np.random.randint(0,n)
        y = np.random.randint(0,n)
        z = np.random.randint(0,n)
        obstacle_data[i] = np.array([x,y,z])
        Grid[x][y][z] = np.inf
    return obstacle_data
```

```python
open_set = PriorityQueue()

class Point:
    def __init__(self,point):
        self.x,self.y,self.z = point[0],point[1],point[2]
        neighbours = []
    

def gpoint(point):
    # print(point)
    x,y,z = point[0],point[1],point[2]
    return Grid[x][y][z]

def euclidian_dist(next_point,end_point):
    xn,yn,zn = next_point
    xg,yg,zg = end_point
    return np.sqrt((xn - xg)**2 + (yn - yg)**2 + (zn - zg)**2)


def add_neig(currn_point):
    x,y,z = currn_point
    neighbours = []
    if x in range(0,n-1) and Grid[x+1][y][z] != np.inf:
        neighbours.append((x+1,y,z))
    if y in range(0,n-1) and Grid[x,y+1,z] != np.inf:
        neighbours.append((x,y+1,z))
    if z in range(0,n-1) and Grid[x,y,z+1] != np.inf:
        neighbours.append((x,y,z+1))

    if x in range(1,n) and Grid[x-1][y][z] != np.inf:
        neighbours.append((x-1,y,z))
    if y in range(1,n) and Grid[x,y-1,z] != np.inf:
        neighbours.append((x,y-1,z))
    if z in range(1,n) and Grid[x,y,z-1] != np.inf:
        neighbours.append((x,y,z-1))
    
    return neighbours
    

def recon_path(came_from,current,start):
    path = np.array([current])
    temp = current
    while current in came_from:
        current = came_from[current]
#         print(current)
        path = np.concatenate((path,np.array([current])))
    return path
    
    
def run_astar(start,end):
    if gpoint(start) == np.inf or gpoint(end) == np.inf:
        print("obstacle point")
        return False
    
    no = 1
    open_set.put((0,0,0,start))
    neighs_list = {start}
    
    came_from = {}
    g_score_data = {start: 0}

    while not open_set.empty():
        
        current = open_set.get()[3]
        neighs_list.remove(current)
        
        if current == end:
            path = recon_path(came_from,current,start)
            print("reached")
            return path
            break
        
        neighbours = add_neig(current)
        
        for neighbour in neighbours:
            
            temp_g_score = g_score_data[current] + 1
            
            
            if g_score_data.get(neighbour) == None:
                g_score_data[neighbour] = float("inf")
        
            
            if temp_g_score < g_score_data[neighbour]:
                came_from[neighbour] = current
                g_score_data[neighbour] = temp_g_score
                h_score = euclidian_dist(neighbour,end)
                f_score = temp_g_score + h_score
                if neighbour not in neighs_list:
                    open_set.put((f_score,temp_g_score,no,neighbour))
                    neighs_list.add(neighbour)
                    no += 1

    


obstacle_data = create_obstacle()
# print(obstacle_data)
path = run_astar((4,5,8),(7,4,7))
# print(path)
# print(open_set.get())
```

```python
fig = plt.figure(figsize=(11,11))

ax = fig.add_subplot(111, projection='3d')

x_data = obstacle_data[:,0]
y_data = obstacle_data[:,1]
z_data = obstacle_data[:,2]
ax.scatter(grid_data[:,0],grid_data[:,1],grid_data[:,2],color= "skyblue",marker='.')
ax.scatter(x_data,y_data,z_data, color='black',marker='D')
zline = path[:,2]
xline = path[:,0]
yline = path[:,1]
ax.plot3D(xline, yline, zline, 'blue')

ax.scatter(0,0,0, color='yellow')
plt.show()
```

```python
class path_panning:
    def __init__(self,n):
        self.n = n
        self.open_sets = []
        self.neigh_dicts = []
        self.came_froms = []
        self.g_score_datas = []
        self.f_score_datas = []
        self.currents = []
        self.neighbours_s = []
        self.same_neighbours = {}
        self.paths = []
    
    def set_data(self,end_points):
        self.end_points = end_points
        for points in end_points:
            open_set = PriorityQueue()
            open_set.put((0,0,0,points[0]))
            self.open_sets.append(open_set)
            self.neigh_dicts.append({points[0]})
            self.came_froms.append({})
            self.neighbours_s.append([])
            self.g_score_datas.append({points[0]: 0})
            self.f_score_datas.append({points[0]: 0})
    
    def recon_path(self,came_from,current):
        path = np.array([current])
        temp = current
        while current in came_from:
            current = came_from[current]
            path = np.concatenate((path,np.array([current])))
        return path 
    
    def euclidian_dist(self,next_point,end_point):
        xn,yn,zn = next_point
        xg,yg,zg = end_point
        return np.sqrt((xn - xg)**2 + (yn - yg)**2 + (zn - zg)**2)
    
    def add_neig(self,currn_point):
        x,y,z = currn_point
        neighbours = []
        if x in range(0,n-1) and Grid[x+1][y][z] != np.inf:
            neighbours.append((x+1,y,z))
        if y in range(0,n-1) and Grid[x,y+1,z] != np.inf:
            neighbours.append((x,y+1,z))
        if z in range(0,n-1) and Grid[x,y,z+1] != np.inf:
            neighbours.append((x,y,z+1))

        if x in range(1,n) and Grid[x-1][y][z] != np.inf:
            neighbours.append((x-1,y,z))
        if y in range(1,n) and Grid[x,y-1,z] != np.inf:
            neighbours.append((x,y-1,z))
        if z in range(1,n) and Grid[x,y,z-1] != np.inf:
            neighbours.append((x,y,z-1))
        return neighbours
        
    def get_low_cost_node(self):
        currents = []
#         for i,open_set in zip(range(self.n),self.open_sets):
#             c = open_set.get()[3]
#             currents.append(c)    
        while len(currents) < self.n:
            
            for i,open_set in zip(range(self.n),self.open_sets):
                if not open_set.empty():
                    c = open_set.get()[3]
                    if c not in currents:
                        currents.insert(i,c)
                    else:
                        pq = [k for k,d in enumerate(currents) if d==c]
                        temp_g_data = []
                        for j in pq:
                            temp_g_data.append(self.g_score_datas[j][c])
                        if self.g_score_datas[i][c] not in temp_g_data:
                            currents.insert(i,c)
#                 print(currents,i) 
        self.currents = currents
                    
    
    def remove_neigh_list(self):
        for current,neigh_list in zip(self.currents,self.neigh_dicts):
#             print(neigh_list)
            neigh_list.remove(current)
            
    def get_neighbours(self):
        i = 0
        for current in self.currents:
            neighbours = self.add_neig(current)
            self.neighbours_s[i] = neighbours
            i +=1
    
    def close_iteration(self,i):
#         print(i)
        self.n -= 1
        self.open_sets.pop(i)
        self.neigh_dicts.pop(i)
        self.came_froms.pop(i)
        self.g_score_datas.pop(i)
        self.currents.pop(i)
#         print(self.currents)
        self.neighbours_s.pop(i)
#         print(self.neighbours_s)
        
    def check_point_reached(self):
        for i,came_from,current in zip(range(self.n),self.came_froms,self.currents):
            if current == self.end_points[i][1]:
                print("reached")
                print(current,self.end_points[i][1])
                path = self.recon_path(came_from,current)
                
                self.paths.append(path)
                self.end_points.pop(i)
                self.close_iteration(i)
    
    def check_same_gscore(self,neighbour,g_score):
        flag = 0
        self.same_neighbours[neighbour] = []
        for g_score_data in self.g_score_datas:
            if g_score_data.get(neighbour):
                if g_score_data[neighbour] == g_score:
                    flag +=1
                    i_n = self.g_score_datas.index(g_score_data)
                    self.same_neighbours[neighbour].append(i_n)
        
    def run_astar(self):
        no = 1
        while self.open_sets:
#             print(no)
            self.get_low_cost_node()
            self.remove_neigh_list()
            self.check_point_reached()
            self.get_neighbours()
#             print(self.neighbours_s)
            for i in range(self.n):
                
                for neighbour in self.neighbours_s[i]:
                    temp_g_score = self.g_score_datas[i][self.currents[i]] + 1
                    if self.g_score_datas[i].get(neighbour) == None:
                        self.g_score_datas[i][neighbour] = float("inf")

                    if temp_g_score < self.g_score_datas[i][neighbour]:
                        self.came_froms[i][neighbour] = self.currents[i]
                        self.g_score_datas[i][neighbour] = temp_g_score
                        h_score = self.euclidian_dist(neighbour,self.end_points[i][1])
                        f_score = temp_g_score + h_score
                        self.f_score_datas[i][neighbour] = f_score
                        self.check_same_gscore(neighbour,temp_g_score)
                        if neighbour not in self.neigh_dicts[i]:
                            self.open_sets[i].put((f_score,temp_g_score,no,neighbour))
                            self.neigh_dicts[i].add(neighbour)
                            no += 1
                        
            
          
p1 = path_panning(3)
uinput = [[(0,0,0),(5,6,8)],
          [(1,3,1),(5,6,7)],
          [(2,3,4),(7,4,7)],
          [(4,5,8),(7,4,8)]]
p1.set_data(uinput)
p1.run_astar()
# print(p1.paths)
```

```python
fig1 = plt.figure(figsize=(11,11))

ax1 = fig1.add_subplot(111, projection='3d')

x_data = obstacle_data[:,0]
y_data = obstacle_data[:,1]
z_data = obstacle_data[:,2]
ax1.scatter(grid_data[:,0],grid_data[:,1],grid_data[:,2],color= "skyblue",marker='.')
ax1.scatter(x_data,y_data,z_data, color='black',marker='s')
for path in p1.paths:
#     print(path)
    xline = path[:,0]
    yline = path[:,1]
    zline = path[:,2]
    
    ax1.plot3D(xline, yline, zline, 'blue')

ax1.scatter(0,0,0, color='yellow')
plt.show()
```

Serpentine velocity control


```python
class snake:
    def __init__(self):  
        self.A_max = 60*np.pi/180
        self.l = 1
        self.cycles = 100
        self.c = 3
        self.V_des = 5
        self.w_ideal = 2*np.pi* self.V_des / ( self.l - self.l*np.cos(self.A_max/2))
        self.v_actual = 0
        self.data_v = np.array([self.v_actual])
        self.data_w = np.array([self.w_ideal])

    def equn(self,omega,t):
        delta = 1.54
        return self.A_max*np.sin(omega*t + delta)

    def get_v_actu(self):
        self.v_actual = self.v_actual + self.V_des*0.1

    def w_for_new_cycle(self):
        self.get_v_actu()
        self.data_v = np.append(self.data_v,self.v_actual)
        self.w_ideal = self.w_ideal + self.c*(self.V_des - self.v_actual)
        

    def plot(self):    
        while self.V_des - self.v_actual > 0:
            self.w_for_new_cycle()
            self.data_w = np.append(self.data_w,self.w_ideal)

s = snake()
s.plot()
t = np.linspace(0,len(s.data_w),11)

plt.plot(t,s.data_w)
plt.show()
```

```python
plt.plot(t,s.data_v)
V_des = [s.V_des for _ in range(len(s.data_w))]

plt.plot(t,V_des,'--')
plt.show()
```
